```python
#4=
from sklearn.preprocessing import LabelEncoder

# Sample data
color_data = ["red", "green", "blue", "green", "blue"]
size_data = ["small", "medium", "large", "medium", "small"]
material_data = ["wood", "metal", "plastic", "metal", "wood"]

# Create instances of LabelEncoder for each categorical variable
color_encoder = LabelEncoder()
size_encoder = LabelEncoder()
material_encoder = LabelEncoder()

# Fit and transform each categorical variable using its respective encoder
encoded_color = color_encoder.fit_transform(color_data)
encoded_size = size_encoder.fit_transform(size_data)
encoded_material = material_encoder.fit_transform(material_data)

# Display the encoded data
print("Encoded Color:", encoded_color)
print("Encoded Size:", encoded_size)
print("Encoded Material:", encoded_material)
```
